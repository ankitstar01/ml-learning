```
import numpy as np
import math
import time
import pandas as pd
```

```
cd=0
ld=0
def get_distance(p1,p2):
    x=p2[0]-p1[0]
    y=p2[1]-p1[1]
    cd=math.sqrt(x*x+y*y)
    return cd
```

```
q=[]
def get_allDistances(t):
    n = len(t)
    e=0
    distances = [[0]*n]*n
    t1=time.time()
    distances=np.array(distances)
    for i in range(n):
        for j in range(n):
            q.append(get_distance(t[i],t[j]))
            e=e+1
            distances[i][j]=get_distance(t[i],t[j])
    t2=time.time()
    print('total time taken {}sec:'.format(t2-t1))
    return distances
```
